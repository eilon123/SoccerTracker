```python
import cv2
import numpy as np
from soccer_tracker.track_models.sort import Sort

COLORS = np.random.randint(0, 255, size=(400, 3))
mot_tracker = Sort()
```

```python
def calculate_iou(bb1, bb2):
    """
    Computes IUO between two bboxes in the form [x1,y1,x2,y2]
    """
    xx1 = np.maximum(bb1[0], bb2[0])
    yy1 = np.maximum(bb1[1], bb2[1])
    xx2 = np.minimum(bb1[2], bb2[2])
    yy2 = np.minimum(bb1[3], bb2[3])
    w = np.maximum(0., xx2 - xx1)
    h = np.maximum(0., yy2 - yy1)
    wh = w * h
    iou = wh / ((bb1[2] - bb1[0]) * (bb1[3] - bb1[1])
              + (bb2[2] - bb2[0]) * (bb2[3] - bb2[1]) - wh)
    return iou

def get_area(box):
    return np.abs((boxes[0][1][0]-boxes[0][1][2])*(boxes[0][1][1]-boxes[0][1][3]))
```

```python
def postprocess(boxes, scores, labels):
    boxes, scores, labels = boxes[0], scores[0], labels[0]
    relevent_indices = []

    num_boxes = len(boxes)
    for cur_index in range(num_boxes):
        cur_area = get_area(boxes[cur_index])
        if scores[cur_index]<0.1 or cur_area<375 or cur_area>12600:
            pass
        best_index = cur_index
        
        for relevent_index in relevent_indices:
            iou = calculate_iou(boxes[best_index], boxes[relevent_index])
            if iou >= 0.5:
                if labels[best_index] != 0 and labels[relevent_index] == 0:
                    best_index = relevent_index
                elif scores[best_index] < scores[relevent_index]:
                        best_index = relevent_index
        
        if best_index == cur_index:
            relevent_indices.append(best_index)
            
    relevent_indices = np.array(relevent_indices)
    
    return boxes[relevent_indices][np.newaxis], scores[relevent_indices][np.newaxis], labels[relevent_indices][np.newaxis]
```

```python
out = cv2.VideoWriter('track.avi', cv2.VideoWriter_fourcc(*'DIVX'), 15, (1920, 1080))

results_generator = data_handler.detections_generator()
for image, (boxes, scores, labels) in results_generator:
    print(len(boxes[0]))
    boxes, scores, labels = postprocess(boxes, scores, labels)
    print(len(boxes[0]))
    track_bbs_ids = mot_tracker.update(boxes[0])
    track_bbs_ids = track_bbs_ids.astype(np.int)

    # draw a bounding box rectangle and label on the image
    for tr in track_bbs_ids:
        color = [int(c) for c in COLORS[tr[-1]]]
        x1 = tr[0]
        y1 = tr[1]
        x2 = tr[2]
        y2 = tr[3]
        cv2.rectangle(image, (x1, y1), (x2, y2), color, 2)
        cv2.putText(image, str(tr[-1]), (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    out.write(image)

out.release()
```
